```python
def getFolders(path):
    """Gets all the folders that are not hidden from a given directory.
    Parameters:
        path: The directory that contains your folders containing .npy files
    Returns:
        A list of all folders that are not hidden from input directory. 
    """
    folders = []
    for folder in listdir(path):
        if not folder.startswith('.'):
            folders.append(path + '/' + folder)
    return folders
```

```python
def getData(folders, index):
    """Gets the numpy arrays from a specific folder
    Parameters:
        folders: A list of folders each containing .npy files
        index: The index number of the folder you want to retrieve, i.e. index = 3 for 'Row_03...'
    Returns:
        A dictionary with .npy file name as key and the array as value. 
    """
    data = {} #Create an empty dictionary to put data from DATAFILES in
    DATAFILES = glob.glob(folders[index] + '/*.np[yz]')
    for np_name in DATAFILES:
        data[np_name] = np.load(np_name)
    return data
```
